### 9) Defining X  and
### 10) Defining Y

```python
training_features = ['age', 'balance', 'campaign', 'pdays', 'previous', 'poutcome_other', 'poutcome_success', 'poutcome_unknown']
X=dat_Bank[training_features]
y=dat_Bank[target_feature]
```

```python
dat_Bank
```

### 11) Split Training and testing sets, then validate distribution

```python
from sklearn.model_selection import train_test_split
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=.40, random_state=500)
```

```python
y_train.mean()
```

```python
y_test.mean()
```

### 12) Application of the decision tree model

```python
from sklearn.tree import DecisionTreeClassifier
class_tree = DecisionTreeClassifier(max_depth=3)
class_tree.fit(X_train, y_train)
```

```python
from sklearn.externals.six import StringIO  
from sklearn.tree import export_graphviz
from IPython.display import Image  
import pydotplus
```

```python
dot_data = StringIO()
export_graphviz(decision_tree=class_tree,
                out_file=dot_data,
                filled=True,
                rounded=True,
                proportion=True,
                feature_names = X_train.columns,
                class_names = ['Low', 'High'],
                special_characters=True)
graph = pydotplus.graph_from_dot_data(dot_data.getvalue())  
Image(graph.create_png())
```

### Creating Confusion Matrix and other performance metrics

```python
from sklearn.metrics import confusion_matrix
y_train_pred = class_tree.predict(X_train)
confusion_matrix(y_train,y_train_pred)

#diagonal(ii,jj) is correct rows, diagonal (ij,ji) are wrong, so bottom outbput says 16080 are coorect as per the model
```

```python
from sklearn.metrics import accuracy_score
from sklearn.metrics import precision_score
from sklearn.metrics import recall_score
from sklearn.metrics import f1_score
```

```python
precision = precision_score(y_train, y_train_pred)
recall = recall_score(y_train, y_train_pred)
accuracy = accuracy_score(y_train, y_train_pred)
f_measure = f1_score(y_train, y_train_pred)
print ('precision: {:0.1f}%, recall:{:.1f}%, accuracy: {:0.1f}%, f_measure :{:0.1f}%'.format(100*precision, 100*recall, 100*accuracy, 100*f_measure))
```

```python
pd.crosstab(y_train['purchasecode_yes'], y_train_pred, rownames=['True'], colnames=['Predicted'], margins=True)
```

```python
from sklearn.metrics import accuracy_score
accuracy = accuracy_score(y_true=y_train, y_pred=y_train_pred)
accuracy.round(3)
```

```python
pd.Series(data=class_tree.feature_importances_, index=X_train.columns).sort_values(ascending=False).round(3)
```

#### Calculation of Precision, Recall, and F-Measure

aa = 2369
ba = 256
ab = 30
bb = 57

Precision (a) = aa/classified a = 2369/(2369+256) = 0.90;
Precision (b) = bb/classified b = 57/(30+57) = 0.65

Recall (a) = aa/actual a = 2369/(2369+30) = 0.98;
Recall (b) = bb/actual b = 57/(256+57) = 0.18

F = [(2*Precision*Recall)/(Precision+Recall)]

F(a) = (2*0.90*0.98)/(0.90+0.98) = 0.94;
F(b) = (2*0.65*0.18)/(0.65+0.18) = 0.28
